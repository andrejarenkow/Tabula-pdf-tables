```python
import pandas as pd
import tabula
```

```python
arquivo_pdf = 'http://lproweb.procempa.com.br/pmpa/prefpoa/concursos/usu_doc/edital_063-2023_-_anexo_iii_notas_preliminares.pdf'
```

```python
tabelas = tabula.read_pdf(arquivo_pdf, pages='all', lattice=True)
len(tabelas)
```

```python
tabelao = pd.DataFrame()
for i in tabelas:
  if len(i.columns)>4:
    tabelao = pd.concat([i.T.reset_index().T.replace('\r',' ', regex=True).replace(',','.', regex=True), tabelao])
    
tabelao = tabelao.reset_index(drop=True)
tabelao
```

```python
tabelao.columns = ['Nome', 'Inscricao','Portugues','Legislacao','Legislacao Especifica','Conhecimentos Pedagogicos','Conhecimentos Especificos','Total','tirar','tirar2']
tabelao = tabelao.drop(['tirar','tirar2'], axis=1)
tabelao
```

```python
colunas_numericas = ['Portugues','Legislacao','Legislacao Especifica','Conhecimentos Pedagogicos','Conhecimentos Especificos','Total']
for i in colunas_numericas:
  tabelao[i] = pd.to_numeric(tabelao[i], errors='coerce')

tabelao
```

```python
tabelao = tabelao.reset_index(drop=True)
```

```python
codigos = {
    '74701':'Professor - Anos Iniciais',
    '74702':'Professor - Artes',
    '74703':'Professor - Ciência Químicas, Físicas e Biológicas',
    '74704':'Professor - Educação Física',
    '74705':'Professor - Educação Infantil',
    '74706':'Professor - Filosofia',
    '74707':'Professor - Geografia',
    '74708':'Professor – História',
    '74709':'Professor – Língua Espanhola',
    '74710':'Professor - Língua Inglesa',
    '74711':'Professor - Língua Portuguesa',
    '74712':'Professor - Matemática',
    '74713':'Professor Educação Especial - Habilitação Deficiência Mental',
    '74714':'Professor Educação Especial - Habilitação Deficiência Visual',
    '74715':'Professor Educação Especial - Habilitação Educação de Surdos',
}

tabelao['Inscricao_codigo'] = [x[:5] for x in tabelao['Inscricao'].astype(str)]
tabelao['Cargo'] = tabelao['Inscricao_codigo'].map(codigos)
tabelao = tabelao.dropna(subset='Cargo')
tabelao
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
data = tabelao.value_counts('Cargo').reset_index()
data.columns = ['Cargo','Inscritos']
px.bar(data, x='Cargo', y='Inscritos', text_auto=True)
```

```python
port = tabelao['Portugues']>=4
legisl = tabelao['Legislacao']>=2
lesgil_esp = tabelao['Legislacao Especifica']>=4
conhec = tabelao['Conhecimentos Pedagogicos']>=8
conhec_esp = tabelao['Conhecimentos Especificos']>=20
tot = tabelao['Total']>=40

filtro_passar = port & legisl & lesgil_esp & conhec & conhec_esp & tot
tabelao['Situacao'] = filtro_passar
tabelao['Situacao'] = tabelao['Situacao'].replace({True:'Aprovado', False:'Não aprovado'})
tabelao.reset_index(drop=True, inplace=True)

notas = []
for i in range(len(tabelao)):
  
  if tabelao['Situacao'][i] == 'Aprovado':
    notas.append(tabelao['Total'][i])
  
  else:
    notas.append(0)

tabelao['Total_rank_final'] = notas

tabelao_rankeado = pd.DataFrame()

for i in tabelao['Cargo'].unique():
  filtro_cargo = tabelao['Cargo']==i
  tabelao_filtrado = tabelao[filtro_cargo].sort_values('Total_rank_final', ascending=False).reset_index(drop=True)
  tabelao_filtrado['rank'] = tabelao_filtrado['Total_rank_final'].rank(method='dense', ascending=False)
  tabelao_filtrado = tabelao_filtrado.sort_values('rank')
  tabelao_rankeado = pd.concat([tabelao_filtrado, tabelao_rankeado])
  


tabelao_rankeado = tabelao_rankeado.reset_index(drop=True)
tabelao_rankeado.to_excel('tabelao_rankeado.xlsx', index=False)
tabelao_rankeado

```
